# ECOD

```python
from pyod.models.ecod import ECOD
```

```python
ana_sc=ECOD(contamination=0.22)
```

```python
st_time=time.time()
ana_sc.fit(data)
end_time=time.time()
```

```python

print(end_time-st_time)
```

```python
ecod_score_nparray=ana_sc.decision_scores_
```

```python
ecod_score=list(ecod_score_nparray)
```

```python
ecod_score
```

```python
ecod_lab=ana_sc.labels_
```

```python
ecod_labels=list(ecod_lab)
```

```python
ecod_labels
```

```python
ecod_threshold=ana_sc.threshold_
```

```python
ecod_threshold
```

### Visualize

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
model = TSNE(n_components = 2, random_state = 0)
```

```python
st_time=time.time()
tsne_data = model.fit_transform(data)
end_time=time.time()
```

```python
print(end_time-st_time)
```

```python
tsne_data1 = np.vstack((tsne_data.T, ecod_labels)).T
tsne_df = pd.DataFrame(data = tsne_data1,
     columns =("Dim_1", "Dim_2", "label"))
```

```python
import seaborn
from matplotlib import pyplot as plt
```

```python
seaborn.FacetGrid(tsne_df, hue ="label", height = 6).map(
       plt.scatter, 'Dim_1', 'Dim_2').add_legend()
 
plt.show()
```

```python
plt.savefig(r'C:\Users\91852\OneDrive\Desktop\anamoly_outliers.png')
```

# Scaling


```python
mx=max(ecod_score)
mn=min(ecod_score)
mean=np.mean(ecod_score)
```

```python
norm_data=((ecod_score-mn)/(mx-mn))*100
```

```python
min(norm_data)
```

```python
max(norm_data)
```

```python
norm_threshold=(ecod_threshold-mn)/(mx-mn)*100
```

```python
norm_labels=[]
noise=0
for i in norm_data:
    if(i<norm_threshold):
        norm_labels.append(0)
    else:
        norm_labels.append(1)
        noise=noise+1

```

```python
noise
```

```python
tsne_data2 = np.vstack((tsne_data.T, ecod_labels)).T
tsne_df = pd.DataFrame(data = tsne_data2,
     columns =("Dim_1", "Dim_2", "label"))
```

```python
seaborn.FacetGrid(tsne_df, hue ="label", height = 6).map(
       plt.scatter, 'Dim_1', 'Dim_2').add_legend()
 
plt.show()
```

```python
plt.savefig(r'C:\Users\91852\OneDrive\Desktop\anamoly_outliers_scaled.png')
```
