### Changing to Normal distribution

```python
import scipy.stats as stats
import pylab
```

```python
np_ecod=np.array(ecod_score)
```

```python
stats.probplot(np_ecod,plot=pylab);
```

```python
y=[]
for i in (-3,-2,-1,-.5,0,.5,(1/1.2),1,2):
    if(i==0):
        y.append(np.log(np_ecod))
    else:
        y.append((np_ecod**i-1)/i)
    
```

### Mean Normalization

```python
x=ecod_score
y=(x-np.mean(x))/(max(x)-min(x))
nt=(ecod_threshold-np.mean(x))/(max(x)-min(x))
```

```python
y=y*100
nt=nt*100
```

```python
m=min(y)
m
```

```python
max(y)
```

```python
y=y-m
nt=nt-m
```

```python
min(y)
```

```python
stats.probplot(y,plot=pylab);
```

```python
plt.plot(sorted(y))
```

```python
max(y)
```

```python
nt
```

```python
norm_labels=[]
noise=0
for i in y:
    if(i<nt):
        norm_labels.append(0)
    else:
        norm_labels.append(1)
        noise=noise+1

```
